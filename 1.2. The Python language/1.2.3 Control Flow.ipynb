# 1.2.3.1 if/elif/else

```python
#Escribo una condicional con una operacion matematica de igualdad

if 2*2 == 4:
    print("Obviuos!")
```

Los bloques de codigo se encuentr DELIMITADOS por IDENTACION

Luego de utilizar el signo dos puntos (:) el shell de Python automaticamente genera la identacion

```python
a = 10 #Defino una variable con valor 10

if a == 1: #1° condicional
    print(1) #Resultado en caso que el 1° condicional sea TRUE. Si el resultado es FALSE continua con la sentencia de ELIF
elif a == 2: #2° condicional
    print(2) #Resultado en caso que el 2° condicional sea TRUE. Si el resultado es FALSE continua con la sentencia de ELSE
else: #3° condicional
    print("A lot!") #Resultado del 3° condicional
```

La identacion es OBLIGATORIA

# 1.2.3.2 for/range

Iterando por medio de un indice:

```python
#En este caso se itera sobre el rango 0-4 y como resultado se imprime el valor iterado

for i in range(4): #Sentencia de iteracion
    print(i) #Resultado
```

Muchas veces es de mejor lectura iterar sobre valores:

```python
for word in ("cool","powerfull","readable"):
    print("Python is", word)
```

# 1.2.3.3 while/break/continue

Ejemplo N°1

```python
z = 1 + 1j
while abs(z) < 100:
    z = z**2 + 1
```

```python
z
```

Ejemplo N°2

```python
z = 1 + 1j

while abs(z) < 100:
    if z.imag == 0:
        break
    z = z**2 + 1
```

```python
z
```

Ejemplo N°3

```python
a = [1,0,2,4]

for element in a:
    if element == 0:
        continue
    print(1. / element)
```

# 1.2.3.5 Advanced iteration

Se puede iterar sobre cualquier secuencia (strings, listas, claves en un diccionario, lineas en un archivo,etc):

```python
vowels = "aeiouy" #Defino la condicion inicial que luego sera utilizada para comparar

for i in "powerful": #Itero sobre cada letra de la palabra que me interesa, en este caso powerful
    if i in vowels: #Si alguna de las letras se encuentra dentro de la variable vowels definida previamente entonces se imprime
        print(i)
```

```python
message = "Hello how are you?" #Asigno a una variable un string
message.split() #Divido el sting en palabras separadas
```

```python
for word in message.split(): #Itero sobre cada palabra del string y las imprimo
    print(word)
```

Keeping track of enumeration number

```python
words = ("cool","powerful","readable")
for i in range(0,len(words)):
    print((i,words[i]))
```

```python
for index, item in enumerate(words):
    print((index,item))
```

Looping over a dictionary

```python
d = {"a":1,"b":1.2,"c":1j}

for key,val in sorted(d.items()):
    print("Key: %s has value: %s" % (key,val))
```
